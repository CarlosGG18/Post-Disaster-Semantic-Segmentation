```python
import streamlit as st
import os
from PIL import Image
import random
import numpy as np
```

```python
def image_out(image):
  return Image.open(image)
```

```python
def stream_run():
  st.title('Post Disaster Semantic Segmenatation')
  st.text("")
  st.text("")
  st.sidebar.markdown('''
  # This project: 
  - Uses a high resolution image dataset after Hurrican Harvey assesing the damages on Texas and Liousinna
  - The images on the right displays the Semantic Segmentations predictions generated by the Unet and PSPNet models
  ''')
  st.sidebar.markdown('''Check out my Github for code and models [here](https://github.com/CarlosGG18/Post-Disaster-Semenatic-Segmentation)''')



  image = ""
  if st.button('Generate'):
    image_n = random.randint(0, len(os.listdir('Data/st_images')))
    image = f"Data/st_images/'{image_n}.png"
    st.image(np.array(image_out(image))[:-30, 100:-80, :], use_column_width=True) 
  if __name__ == "__main__":
    st.set_page_config(
    initial_sidebar_state="expanded",
    page_title="Post-Disaster Semantic Segmentation"
    )
    stream_run() 
```

```python
stream_run()
```
